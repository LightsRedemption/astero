```python
def noiserv(numax, dnu, time): 
    rv = Amp * np.sin(2 * np.pi * (numax + dnu )  * time)
    
    #currently trying to add noise to data
    sigma = 0.001
    mu =0 
    noise = np.random.normal(mu,sigma,rv.shape)
    noise_rv = rv + noise

def time_hr(time):
        time_hours = time * 1e6 / (3600) #with 0.1 = 27.78 hours
        return time_hours
    
def freq_plot(time, noise_rv):
    
    frequency, power = LombScargle(time, noise_rv).autopower(normalization = 'psd')
    return frequency, power

    #originally had this....but the orginial plot without the axis change doesnt normally look like what we are seeing
    #plt.axis([0,100 , 0, 1.05])
    #plt.xlabel('Frequency')
    #plt.ylabel('Power')
    #plt.title('Power Spectrum')
    #plt.plot(frequency, power)
    #plt.show()


def osc_plot(time_hours, noise_rv):
    
    start = 43
    end = 86
    
    if (len(time_hours) >= 100 ):
        plt.plot(time_hours, noise_rv, label = 'Continuous', color = 'Black' )
        #plt.plot(time_hours[:start], noise_rv[:start] )
        #plt.plot(time_hours[end:], noise_rv[end:])
    else: 
        plt.scatter(time_hours, noise_rv, label = f'{len(time_hours)} Samples', color = np.random.rand(3,))
        
    plt.title('Oscillations')
    plt.xlabel('Time (hours)')
    plt.ylabel('RV $\mathregular{kms^{-1}}$')
    plt.legend(loc = 'best')
    #plt.show()

def radial_velocity(time):
    noiserv(numax, i, time_hr(time))
    osc_plot(time_hr(time), noiserv(numax,i,time))
    freq_plot(time, noiserv(numax,i,time))
    #freq_plot(time[:43], noiserv(numax,time[:43]))
    
def random_time(time): 
    time_array = []
    for i  in range(len(time)):
        value = random.uniform(min(time),max(time)) 
        time_array.append(value)
    time_array.sort()
    return np.array(time_array)
    
```

```python
def grgb(nu, T):
    g_rbg = logg(nu, T, emp = True) #g, surface gravity of red giant, T = 4480 +- 62 
    return 10**g_rbg
def Mrgb(nu, T, R, sin_inc):
    g = grgb(nu,T)
    return ((g * (R*R_sun)**2)/G)/ ( M_sun ) #/ sin_inc**2
```

```python
def Mco(Df):
#first setting inclination values to match with Thompson paper and error of 0.8 +- 0.2
    fig , ax = plt.subplots(2, figsize=(8,6))
    plt.tight_layout()
    for i in range(5):
        Df['sini'] = 0.6 + 0.1 * i
        M_rgb = []
        M_co = []
#begin looping over dataframe rows to create different M_rgb arrays to compute M_co arrays 
        for (Nu,R,T, F, sin_inc) in zip(Df['nu'], Df['R'], Df['T'], Df['F'], Df['sini']):
            mrgb = Mrgb(Nu,T,R,sin_inc)
            sol = root_scalar(f, args = (sin_inc, F, mrgb), x0 = 5.0,fprime = fderiv, maxiter = 10000)
            #only taking converged values, doesn't seem to change smoothing though
            if sol.converged: 
                M_rgb.append(mrgb)
                M_co.append(sol.root)
        M_rgb = np.array(M_rgb)
        M_co = np.array(M_co)
        sort = np.argsort(M_rgb)
        M_rgb = M_rgb[sort]
        M_co = M_co[sort]
#determining 5th and 95th percentile ranges to use as an overlay plot
        M_co95 = np.percentile(M_co, 95)
        M_rgb95 = np.percentile(M_rgb,95)
        M_co5 = np.percentile(M_co, 5)
        M_rgb5 = np.percentile(M_rgb,5)
#plotting stuff
        ax[0].plot(M_rgb, M_co, color = 'black')
        ax[0].set_xlim(0,6)
        ax[0].set_xticks(np.arange(0,7))
        ax[0].set_yticks(np.arange(0,9))
        ax[0].set_ylim(0.75,8)
        ax[0].plot(M_rgb95,M_co95, marker = '|', markersize = '60', linewidth = '3', color = 'red' )
        ax[0].plot(M_rgb5, M_co5, marker = '|', markersize = '55', linewidth = '3', color = 'red')
        M_rgbf = M_rgb[np.where((M_rgb5 <= M_rgb) & (M_rgb <= M_rgb95))]
        M_cof = M_co[np.where((M_co5 <= M_co) & (M_co <= M_co95))]
        ax[0].plot(M_rgbf, M_cof, color = 'r')
        ax[0].grid(True)
        ax[0].set_xlabel(r'$M_{\mathrm{rgb}}$')
        plt.ylabel(r'$M_{\mathrm{co}}$')
        plt.title('Compact Object Mass at Varying Inclincations' )
        print ('Mean Compact Object Mass:' , np.mean(M_co))
        #sns.histplot(x=M_rgb, y=M_co, bins=50, pthresh=.1, cmap="mako", ax = ax[1])        
        #sns.scatterplot(x=M_rgb, y=M_co, ax = ax[1])
        #sns.kdeplot(x=DF['sini'], y=M_co, levels=5, color="w", linewidths=1, ax =  ax[1])
        
```

```python
#creating histogram function to create M_co sampling histograms 
def mco_hist(DF):
    M_rgb = []
    M_co = []
    for (Nu,R,T, F, sin_inc) in zip(DF['nu'], DF['R'], DF['T'], DF['F'], DF['sini']):
        mrgb = Mrgb(Nu,T,R,sin_inc)
        root = root_scalar(f, args = (sin_inc, F, mrgb), x0 = 5.0, fprime = fderiv, maxiter = 10000).root
        M_rgb.append(mrgb)
        M_co.append(root)
    M_rgb = np.array(M_rgb)
    M_co = np.array(M_co)
    sort = np.argsort(M_rgb)
    M_rgb = M_rgb[sort]
    M_co = M_co[sort]
    m = np.mean(M_co)
    plt.hist(M_co, label = f'{len(DF)}', histtype = 'step', density = True)
    plt.axvline( m , color = 'k', linestyle = 'dashed')
    plt.xlabel('Mass of Compact Object')
    plt.title('Samplings of Compact Object Mass') 
    plt.legend()
```
